둘간의 오차를 줄이기

하나의 

$$ \hat{{y}_{i,j}} = a_{i,j} \cdot p_{만화특성} + b_{i,j} \cdot q_{유저특성} $$

여기서 $i$ 는 하나의 privacy_1이며, $j$는 하나의 comic_hash임

$$ matrix\ \hat{Y} = A \cdot P + B \cdot Q $$

```python
class user_comic(object):
    def __init__(self, fileloc):
        column1 = ['buyinhour', 'plat_A', 'plat_B', 'plat_C', 'plat_D', 'total_session',\
                   'comic_hash', 'privacy_1', 'privacy_2', 'privacy_3'] 
        column2 = ['comic' + str(x+1) for x in range(100)] 
        column3 = ['comic_tag1', 'coin_needed', 'end']
        column4 = ['schedule' + str(x+1) for x in range(123-114+1)]
        column5 = ['genre' + str(x+1) for x in range(141-124+1)] 
        column6 = ['last_episode', 'book', 'comic_start', 'total_episode']
        column7 = ['comic_tag' + str(x+2) for x in range(151-146+1)] 
        column8 = ['user_tendency' + str(x+1) for x in range(167-152+1)]
        columns = column1 + column2 + column3 + column4 + column5 + column6 + column7 + column8
        fileloc = './data/lezhin_dataset_v2_training.tsv'
        data = pd.read_csv(fileloc, sep='\t', header=None, names=columns )
        
        # columns that i want to use
        column_new = ['buyinhour', 'pirvacy_1'] + column2 + ['comic_hash', 'plat_A', 'plat_B', 'plat_C', 'plat_D', \
                      'total_session', 'coin_needed', 'end'] + column4 + column5 + column6
        self.data_ = data.loc[:, column_new]
        self.column_user = ['pirvacy_1'] + column2 
        self.column_comic = ['comic_hash', 'plat_A', 'plat_B', 'plat_C', 'plat_D', 'total_session', 'coin_needed', 'end', \
                       ] + column4 + column5 + column6
        self.y = self.data_['buyinhour']
        self.X = self.data_.iloc[:, 1:]
        self.user_table = pd.DataFrame(None, index=column2) #transform 해버림, 코드짤때 편해서
        self.comic_table = pd.DataFrame(None, index=self.column_comic[1:])
        
    def train(self):
        # user purchasing power(유저특성, top 100구매이력)
        user_data = self.data_.loc[:, self.column_user].values
        comic_data = self.data_.loc
        for i in range(len(self.data_)):
            user = user_data[i,0]
            if user in self.user_table.columns:
                self.user_table[user] = np.maximum(self.user_table[user].values, user_data[i][1:])
            else:
                self.user_table[user] = user_data[i][1:]
            
            

            
            
            
```
